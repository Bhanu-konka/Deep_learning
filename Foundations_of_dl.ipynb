```python
import numpy as np

# OR gate inputs
x = np.array([
    [1,0],
    [0,1],
    [0,0],
    [1,1]
])

# OR gate outputs
y = np.array([1,1,0,1])

# initial weights and bias
w = np.array([0.1,0.1])
b = 0.1

eta = 0.1   # learning rate

# training
for epoch in range(1,6):

    print(f"\nEpoch {epoch}")

    for i in range(len(x)):

        net_input = np.dot(x[i], w) + b

        # step activation
        if net_input > 0.5:
            y_pred = 1
        else:
            y_pred = 0

        error = y[i] - y_pred

        # update rule
        w = w + eta * error * x[i]
        b = b + eta * error

        print("Input:", x[i])
        print("Pred:", y_pred)
        print("Error:", error)
        print("Updated weights:", w)
        print("Updated bias:", b)
```

```python
# NAND gate using single layer perceptron

import numpy as np

# Inputs
x = np.array([
    [1,0],
    [0,1],
    [1,1],
    [0,0]
])

# NAND outputs
y = np.array([1,1,0,1])

# Initialize weights and bias
w = np.array([0.1, 0.1])
b = 0.1

# learning rate
eta = 0.1

# Training
for epoch in range(10):

    print(f"\nEpoch {epoch}")

    for i in range(len(x)):

        print("Input:", x[i])

        # net input
        net_input = np.dot(x[i], w) + b

        # step activation
        if net_input >= 0:
            y_pred = 1
        else:
            y_pred = 0

        # error
        error = y[i] - y_pred

        # update rule
        w = w + eta * error * x[i]
        b = b + eta * error

        print("Predicted:", y_pred)
        print("Error:", error)
        print("Updated weights:", w)
        print("Updated bias:", b)
```
